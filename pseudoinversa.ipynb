```python
import numpy as np

np.set_printoptions(suppress=True)
```

```python
A = np.array([[1,2], [3,4],[5,6]])
print(A)
```

```python
U, D, V = np.linalg.svd(A)
print('U')
print(U)
print('D')
print(D)
print('V')
print(V)
```

```python
D_pse = np.zeros((A.shape[0], A.shape[1])).T
print(D_pse)
```

```python
print('Valores_a_remplazar_en_D_pse')
print(D_pse[:D.shape[0], : D.shape[0]])
```

```python
print('Valores_que_pondremos_en_D_pse')
print(np.linalg.inv(np.diag(D)))
```

```python
print('D_pse')
D_pse[:D.shape[0], :D.shape[0]] = np.linalg.inv(np.diag(D))
print(D_pse)
```

```python
A_pse = V.T.dot(D_pse).dot(U.T)
print(A_pse)
```

```python
A_pse_calc = np.linalg.pinv(A)
print(A_pse_calc)
```

```python
print(A_pse.dot(A))
```
