```python
import numpy as np
```

```python
np.set_printoptions(suppress=True)
```

```python
A = np.array([[3,1],[2,1]])
print(A)
```

```python
b = np.array([[1],[1]])
```

```python
inversa_A = np.linalg.inv(A)
print(inversa_A)
```

```python
x = inversa_A.dot(b)
print(x)
```

```python
print(A.dot(x))
```

```python
sol_2=inversa_A.dot(np.array([[3],[7]]))
sol_2
```

```python
print(A.dot(sol_2))
```
