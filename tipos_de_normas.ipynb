```python
import numpy as np
```

### Norma 0

```python
vector1 = np.array([1,2,0,5,6,0])

print(np.linalg.norm(vector1, ord = 0))
```

### Norma 1

```python
vector2 = np.array([1,-1,1,-1,1])
print(np.linalg.norm(vector2, ord = 1))
```

### Norma 2

```python
vector3= np.array([1,1])
```

```python
print(np.linalg.norm(vector3))
print(np.linalg.norm(vector3, ord=2))
```

```python
vector4 = np.array([1,2,3,4,5,6])
print(vector4)
```

### Norma mas usada es la L2**2


```python
print(np.linalg.norm(vector4, ord=2)**2)
```

```python
print(vector4.T.dot(vector4))
```

### Norma Infinito

```python
vector5 = np.array([1,2,3,-100])
print(np.linalg.norm(vector5, ord=np.inf)) # np.inf epresenta la norma infinita en numpy
```
