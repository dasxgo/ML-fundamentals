```python
import numpy as np
```

```python
A = np.array([[2,3],[5,7],[11,13]])
B = np.array([[1,3],[2,1]])
C = np.array([[3,1],[4,2]])
```

### Asociativa

A x (BXC) = (AXB) XC

```python
ABC = A.dot(B.dot(C))
print(ABC)
```

```python
AB_C = A.dot(B).dot(C)
print(AB_C)
```

### Distributiva

A X (B+C) = (AXB) + (AXC)


```python
D = A.dot(B+C)
E = (A.dot(B))+(A.dot(C))

print(D)
print(E)
```

```python
print(D==E)
```

### Conmutativa

BXC = CXB


```python
F = B.dot(C)
```

```python
G = C.dot(B)
```

```python
print(F==G)
```

```python
print(F)
```

```python
print(G)
```
