```python
import numpy as np
```

### Matriz diagonal


```python
vector = np.array([1,2,3,4,5])

matriz = np.diag(vector)

print(matriz)
```

### Matriz diagonal no cuadradas


```python
print(matriz[0:4,0:3])
```

```python
print(matriz[0:3,0:4])
```

```python
A = np.diag([2,3,4,5])
print(A)
```

```python
v1 = np.array([[1,1,1,1]])
print(v1)

print(A.dot(v1.T))
```

```python
A_inv = np.diag([1/2, 1/3, 1/4, 1/5])
print(A_inv)
```

```python
identidad = A.dot(A_inv)
print(identidad)
```

```python
A_inv_cal = np.linalg.inv(A)
print(A_inv_cal)
```
